### Retrieve Nucleotide Sequences from NCBI Entrez for Gene IDs

```python
from Bio import Entrez
from urllib.error import HTTPError  
Entrez.email = "sarmadjutt136@gmail.com"
gene_ids = ["NM_001256799", "NM_001354284"]
for gene_id in gene_ids:
    try:
        with Entrez.efetch(db="nucleotide", id=gene_id, rettype="fasta", retmode="text") as handle:
            print(handle.read())
    except HTTPError as e:
        print(f"Failed to fetch data for gene ID {gene_id}: {e}")
```
