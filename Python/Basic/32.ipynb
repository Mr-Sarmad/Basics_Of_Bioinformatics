### Sliding window algorithm to calculate GC content in a genome


```python
from Bio import SeqIO

# Function to calculate GC content
def calculate_gc_content(dna_sequence):
    gc_count = dna_sequence.count('G') + dna_sequence.count('C')
    return (gc_count / len(dna_sequence)) * 100

# Function to calculate GC content in sliding windows
def sliding_window_gc_content(dna_sequence, window_size):
    gc_contents = []
    for i in range(len(dna_sequence) - window_size + 1):
        window = dna_sequence[i:i + window_size]
        gc_contents.append(calculate_gc_content(window))
    return gc_contents

# Path to the FASTA file
fasta_file = "C:/Users/PC/Downloads/sequence (3).fasta"

# Extract DNA sequence from FASTA file
for record in SeqIO.parse(fasta_file, "fasta"):
    dna_sequence = str(record.seq)

# Print the sliding window GC content (with window size 5)
print(sliding_window_gc_content(dna_sequence, 5))


```
